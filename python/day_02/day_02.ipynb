## 1. Micro Challenge: The Guard Clause (Short-Circuting)

##### Goal create a variable user = None. write an if statement that checks if user is not None and if user has "admin" access.

##### Constraint: must perform both checks in a single line. it must not crash when user is none.

```python
user = None

if user is not None and user == "admin":
    print("user has admin access.")
else:
    print("user is none.")
```

## 2. Micro Challenge The Floating Point Trap

##### Goal: Write a script that checks if 0.1 + 0.2 == 0.3 print the result (True/False)
##### Observation: it Print False

```python
x = 0.1
y = 0.2

if x+y == 0.3:
    print(True)
else:
    print(False)
    
```

```python
## Fix using tolerance threshold (epsilon)


x = 0.1
y = 0.2

epsilon = 1e-9

''' Why 1e-9 works well in Python

Python floats are 64-bit (IEEE-754 double precision).

They have about 15–16 decimal digits of precision

Errors like 0.1 + 0.2 are around 1e-16

So:

1e-12 → very strict

1e-9 → safe and forgiving

1e-6 → loose

1e-9 sits comfortably above typical floating-point error but below anything meaningful in most calculations.
'''

if abs((x + y) - 0.3) < epsilon:
    print(True)
else:
    print(False)
```
