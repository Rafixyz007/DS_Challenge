## 1. Micro Challenge : The Identity Swap

##### Goal: Initialize x = 100 and y = 200 Swap their values sp x becomes 200 and y becomes 100.

##### Constraint: You must do this in exactly one line of code.

```python
x = 100
y = 200
print(f"Initial value of x = {x} and y = {y}\n")

x,y = y,x
print(f"After swaping value of x = {x} and y = {y}")
```

## 2. Micro Challenge: The Type Auditor

##### Goal: Ask the user to type a number.
##### 1. Print the type() of the raw input.
##### 2. Cast it to a float.
##### 3. Print the type() again


```python
number = input("Enter a number: ") #  By default it takes string
print(type(number))

number = float(number) # Convert string to float (typecasting)
print(type(number))
```

## 3. Micro Challenge: The Precision Bnaker

##### Goal: Print the result formatted strictly to 3 decimal places with a dollar sign

```python
bill = 100 / 3
print(f"${bill:.3f}")
```

## 4. Micro Challenge: The Modulo Architect

##### Goal: input a raw number of seconds use integer division "//" and modulo "%" to calculate exactly how many hours, minutes, and seconds this represents.

```python
number = int(input("enter any integer number"))

hours = number // 3600
minutes = (number % 3600)//60
seconds = number % 60

print(f" number of hours {hours}: number of minutes {minutes}: number of second {seconds}")
```
